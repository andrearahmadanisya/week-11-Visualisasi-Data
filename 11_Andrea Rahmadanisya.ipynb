```python
#Import modules
%matplotlib inline

import matplotlib.pyplot as plt

#for network creation
import networkx as nx
```

```python
G = nx.Graph() #initialize empty graph 
G.add_nodes_from(["Bandung", "Karawang", "Bekasi","Purwakarta","Subang","Soreang","Sumedang", "Indramayu", "Majalengka","Bogor", "Ciamis", "Cianjur","Cimahi", "Cirebon","Garut", "Depok", "Sukabumi"])
G.nodes()
nx.draw(G, with_labels=True)
plt.show()
```

```python
G.add_edges_from([("Bekasi","Karawang"),("Karawang","Purwakarta"), ("Purwakarta","Subang"), ("Purwakarta", "Cimahi"), ("Subang", "Bandung"), ("Bandung", "Soreang"), ("Soreang", "Sumedang"),("Sumedang", "Cirebon"), ("Cirebon", "Indramayu"), ("Indramayu", "Majalengka"), ("Majalengka", "Ciamis"), ("Ciamis", "Tasikmalaya"),
("Tasikmalaya", "Garut"), ("Cianjur", "Bogor"), ("Cianjur", "Sukabumi"), ("Sukabumi","Depok"),("Cianjur","Cimahi")])
nx.draw(G, with_labels = True)
plt.show()
```

```python
list(G.edges(["Bandung","Sukabumi","Bekasi", "Cimahi", "Cianjur"]))
```

```python
list(G.neighbors("Bandung"))
```

```python
plt.figure(figsize=(20,20))
pos = nx.spring_layout(G, k=.1)
nx.draw_networkx(G, pos, node_size=25, node_color='yellow', with_labels=True, edge_color='black')
plt.show()
```

Degree Centrality

```python
degree_centrality = nx.degree_centrality(G)
degree_centrality
```

Betweenness centrality

```python
betweenness = nx.betweenness_centrality(G)
betweenness
```
